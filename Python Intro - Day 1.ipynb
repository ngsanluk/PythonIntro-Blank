# List Comprehension
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

**Build a new student of student who has 'y" in his/her name**:
```
students = ["April", "Ben", "Cathy", "Danny", "Eve"]
newlist = []

for s in students:
  if "y" in s:
    newlist.append(s)

print(newlist)

```

**A much shorter version**:
```
students = ["April", "Ben", "Cathy", "Danny", "Eve"]
newlist = [s for s in students if "y" in s]
print(newlist)
```

**Syntax**:
```
newlist = [expression for item in iterable if condition == True]
```
**Example**:
```
newlist = [s for s in students if len(s)>3]

```

# More List Comprehension

# for loop
- A ```for``` loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
- With the ```for``` loop we can execute a set of statements, once for each item in a list, tuple, set etc.

**Example**:
```
students = ["April", "Ben", "Cathy", "Danny", "Eve"]
for s in students:
    print(s)
```


**Example 2**:
```
s = "ABCDEFG"
for c in s:
    print(c)
```

# for loop with ```break```, ```continue```
- ```break``` statement we can stop the loop before it has looped through all the items
- ```continue``` statement we can stop the current iteration of the loop, and continue with the next

**Example**:
```
students = ["April", "Ben", "Cathy", "Danny", "Eve"]
for s in students:
    if s == "Ben":
        continue
    if s == "Danny":
        break
    print(s)
```

class Account:
    def __init__(self, account_no, balance):
        self.account_no = account_no
        self.balance = balance
        
    def print_account_info(self):
        print("Account Information:")
        print("====================")
        print("A/C #: " + str(self.account_no))
        print("Balance: " + str(self.balance))

ben = Account(3344, 8888.88)
ben.print_account_info()
```

# Exercise
- Define a function name my_func to do the following mathematic calculation
- It accepts three parameters: `a`, `b`, `c`
\\[ 2a^3 + 5b^2 - 8c \\]

```
